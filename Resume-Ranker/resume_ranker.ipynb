Import

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Sample Data

```python
job_desc = "Looking for a data scientist with python, machine learning, and data analysis skills"

resumes = [
    "Experienced in python and machine learning",
    "Web developer with HTML CSS skills",
    "Data analyst skilled in python and statistics",
    "Marketing specialist with communication skills"
]
```

TF-IDF

```python
documents = [job_desc] + resumes

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
```

Similarity

```python
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
```

Ranking

```python
scores = similarity.flatten()

ranked_resumes = sorted(
    list(zip(resumes, scores)),
    key=lambda x: x[1],
    reverse=True
)

for resume, score in ranked_resumes:
    print(f"{resume} → Score: {score:.2f}")
```

SAVE MODEL

```python
import joblib

joblib.dump(vectorizer, "vectorizer.pkl")
```
